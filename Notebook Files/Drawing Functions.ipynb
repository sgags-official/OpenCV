```python
# Drawing a Line

# Python program to explain cv2.line() method 
 
# importing cv2 
import cv2
 
# Reading an image in default mode
image = cv2.imread("C:/Users/Sgags/Downloads/exam.jpg")

# Check if image is loaded properly
if image is None:
    print("Error: Image not found at the specified path.")
    exit()

# Window name in which image is displayed
window_name = 'Image'

# Start coordinate, here (0, 0)
# represents the top left corner of image
start_point = (0, 0)

# End coordinate, here (1200, 800)
# represents the bottom right corner of image
end_point = (1200, 799)

# Green color in BGR
color = (0, 0, 255)

# Line thickness of 9 px
thickness = 9

# Using cv2.line() method
# Draw a diagonal green line with thickness of 9 px
image = cv2.line(image, start_point, end_point, color, thickness)

# Displaying the image 
cv2.imshow(window_name, image)

# Wait until a key is pressed
cv2.waitKey(0)

# Close all windows
cv2.destroyAllWindows()
```

```python
#Circle

import cv2

src = cv2.imread("C:/Users/Sgags/Downloads/waldo.jpg")

# Draw multiple circles
cv2.circle(src, center=(50, 450), radius=20, color=(255, 0, 0), thickness=3)  # Blue circle
cv2.circle(src, center=(500, 300), radius=30, color=(0, 255, 0), thickness=5)  # Green circle
cv2.circle(src, center=(700, 100), radius=40, color=(0, 255, 255), thickness=2)  # Red circle

cv2.imshow("Multiple Circles", src)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Python program to explain cv2.putText() method
  
# importing cv2
import cv2

  
# Reading an image in default mode
image = cv2.imread("C:/Users/Sgags/Downloads/waldo.jpg")
  
# Window name in which image is displayed
window_name = 'Image'

# font
font = cv2.FONT_HERSHEY_SIMPLEX

# org
org = (50, 50)

# fontScale
fontScale = 1
 
# Blue color in BGR
color = (0, 255, 255)

# Line thickness of 2 px
thickness = 2
 
# Using cv2.putText() method
image = cv2.putText(image, 'I dont Pay taxes', org, font, 
                   fontScale, color, thickness, cv2.LINE_AA)
 
# Displaying the image
cv2.imshow(window_name, image)
```

```python
import cv2

# Reading an image in default mode
image = cv2.imread("C:/Users/Sgags/Downloads/batman.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Image not found at the specified path.")
else:
    # Window name in which image is displayed
    window_name = 'Image'

    # Font
    font = cv2.FONT_HERSHEY_SIMPLEX

    # Bottom-left corner of the text
    org = (100, 50)

    # Font scale
    fontScale = 1

    # Blue color in BGR
    color = (0, 255, 255)

    # Line thickness of 2 px
    thickness = 2

    # Using cv2.putText() method
    cv2.putText(image, 'I dont pay Taxes', org, font, fontScale, color, thickness, cv2.LINE_AA)

    # Displaying the image
    cv2.imshow(window_name, image)

    # Wait for a key press and close the window
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2

# Reading an image in default mode
image = cv2.imread("C:/Users/Sgags/Downloads/mona.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Image not found at the specified path.")
else:
    # Window name in which image is displayed
    window_name = 'Image'

    # Center coordinates of the ellipse
    center_coordinates = (400, 250)

    # Length of the ellipse axes
    axesLength = (300, 400)

    # Angle of rotation of ellipse in degrees
    angle = 0

    # Starting angle of the elliptic arc
    startAngle = 0

    # Ending angle of the elliptic arc
    endAngle = 360

    # Red color in BGR
    color = (0, 0, 255)

    # Line thickness of 5 px
    thickness = 5

    # Using cv2.ellipse() method to draw the ellipse
    cv2.ellipse(image, center_coordinates, axesLength, angle, startAngle, endAngle, color, thickness)

    # Displaying the image
    cv2.imshow(window_name, image)

    # Wait for a key press and close the window
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
